```python
import numpy as np

# Making the array for demand
demand = np.random.normal(150,20,1000)

# Convert demand to a list
demand = demand. tolist()

# Convert demand list to int
for i in range(0, len(demand)):
    demand[i] = int(demand[i])

# Accumulators

manufactured = 100

average_profit_list = []

standard_dev_list = []

# To find how many earbuds we should manufacture
while manufactured >= 100 and manufactured < 200:

    # Finding Sales

    # Creating sales list
    sales = [None]*len(demand)

    for i in range(0, len(demand)):
        #If we don't make enough to satisfy the demand then sales will be the 
        #amount we made
        if demand [i] > manufactured:
            sales[i] = manufactured
        #If we can exactly satisfy or make more than the demand
        #sales will equal the quantity demanded
        else:
            sales[i] = demand[i]


    # Finding Revenue

    # Making the list for Revenue
    revenue = [None]*len(sales)

    # Revenue = Sales * Price
    for i in range(0, len(sales)):
        revenue[i] = sales[i]*150


    # Profit = Revenue - Manufacturing Cost - Disposal cost

    # Profit = Revenue - cost


    # Finding Costs

    # Cost = Manufacturing Cost + Disposal cost

    # Finding manufacturing cost
    Manufacturing_cost = manufactured * 28.50

    # Creating list for manufacturing
    Manufacturing_cost_list = [None]*len(sales)

    # The manufacturing cost is the same for every sale
    for i in range(0, len(sales)):
        Manufacturing_cost_list[i] = Manufacturing_cost


    # Finding Disposal Cost
    disposal = [None]*len(demand)

    for i in range(0, len(demand)):
        # If we made enough to satisfy demand we didn't have to dispose 
        #any headphones
        if demand[i] > manufactured:
            disposal[i] = 0
        # If we made more than the demand, we have to take the extra that 
        #we made
        # Then pay $8.50 for each headphone disposed of
        else:
            disposal[i] = (manufactured - demand[i])*8.5


    # Finding Profit

    Profit = [None]*len(revenue)

    for i in range(0, len(revenue)):
        # Profit = Revenue - Manufacturing Cost - Disposal cost
        Profit[i] = revenue[i] - Manufacturing_cost_list[i] - disposal[i]

    # Finding mean of profit list
    
    average_profit = np.around(sum(Profit)/len(Profit),2)

    # Finding standard deviation of profit list

    standard_dev = np.std(Profit)
    
    # Adding to the accumulators
    manufactured += 2

    average_profit_list += [average_profit]
    
    standard_dev_list += [standard_dev]

    # End of loop

# Converting std of max average profit into int

for i in range(0, len(average_profit_list)):
    standard_dev_list[i] = int( standard_dev_list[i])

# Converting average profit into int

for i in range(0, len(average_profit_list)):
   average_profit_list[i] = int(average_profit_list[i])

# Find the max profit
max_profit = max(average_profit_list)

# Finding the quantity that corresponds to the max profit
quantity_pos = average_profit_list.index(max_profit)

# Finding the standard dev that corresponds with maximum profit
std_dev_avg_profit_list = standard_dev_list[quantity_pos]

# Finds the number of how many headphones we should make
quantity_produced = 100 + (quantity_pos*2)

print('Question 1: We should produce ' + str(quantity_produced) + 
      ' headphones to get a maximum profit of $' + str(max_profit) + 
      ' with a standard deviation of $' + str(std_dev_avg_profit_list) + '.')
```
